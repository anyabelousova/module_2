```python
import matplotlib.pyplot as plt
```

```python
sm['score'].value_counts()

def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=sm.loc[objects.loc[:, column].isin(sm.loc[:, column].value_counts().index[:5])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
    
for col in list(objects.columns):
    get_boxplot(col)
    
    
#Отметим, что папина работа учителем положительно влияет на успеваемость учеников, в отличии от другого рода деятельности.
#Желание получить высшее образование значительно влияет на успеваемость учеников в положительную сторону.
    
```

```python
from itertools import combinations
from scipy.stats import ttest_ind

def get_stat_dif(column):
    cols = objects.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(sm.loc[sm.loc[:, column] == comb[0], 'score'], 
                        sm.loc[sm.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
            
for col in columns:
    get_stat_dif(col)
```
